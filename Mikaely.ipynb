```python
# =========================
#   CALCULADORA INTERATIVA
# =========================

while True:
    print("\n=========================")
    print("  CALCULADORA INTERATIVA")
    print("=========================")
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")

    opcao = input("Opção: ")

    # =========================
    #       OPÇÃO 0 - SAIR
    # =========================
    if opcao == "0":
        print("Encerrando o programa... Até a próxima!")
        break

    # =========================
    #   OPÇÃO 1 - MÚLTIPLOS DE 5
    # =========================
    elif opcao == "1":
        print("\n--- Múltiplos de 5 em um Intervalo ---")
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        print(f"\nMúltiplos de 5 entre {inicio} e {fim}:")
        for num in range(inicio, fim + 1):
            if num % 5 == 0:
                print(num)

    # =========================
    #   OPÇÃO 2 - NÚMEROS PARES
    # =========================
    elif opcao == "2":
        print("\n--- Contagem de Números Pares ---")
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        cont = 0
        for num in range(inicio, fim + 1):
            if num % 2 == 0:
                cont += 1

        print(f"\nQuantidade de números pares no intervalo: {cont}")

    # =========================
    #   OPÇÃO 3 - NÚMERO PRIMO
    # =========================
    elif opcao == "3":
        print("\n--- Verificador de Número Primo ---")
        n = int(input("Digite um número inteiro positivo: "))

        if n <= 1:
            print("NÃO é primo (números menores ou iguais a 1 não são primos).")
        else:
            primo = True
            for i in range(2, n):
                if n % i == 0:
                    primo = False
                    break

            if primo:
                print(f"{n} é um número PRIMO!")
            else:
                print(f"{n} NÃO é um número primo.")

    # =========================
    #   OPÇÃO 4 - CONVERSÃO °C → °F
    # =========================
    elif opcao == "4":
        print("\n--- Conversor de Temperatura ---")
        c = float(input("Digite a temperatura em Celsius (°C): "))
        f = c * 1.8 + 32
        print(f"{c:.2f}°C corresponde a {f:.2f}°F")

    # =========================
    #   OPÇÃO INVÁLIDA
    # =========================
    else:
        print("\nOpção inválida! Tente novamente.")
```
